## 3.4 處理一整個list的數字 ==> array 導向

```python
# 美金價格
p = [1096.95,596.95,529.2]
# 匯率
c = 30.2
for k in p:
    print(k*c)
```

```python
result = []
for k in p:
    twd = k*c
    result.append(twd)    
```

```python
result
```

#### 能不用迴圈就不用迴圈
##### 算成績

```python
prices = np.array(p)
prices
```

```python
prices*c
```

```python
grade = np.array([85,72,93])
weight = np.array([0.3,0.4,0.3])
g = grade*weight
print(g)
print(g.sum())
# 內積
np.dot(grade,weight)
```

#### 算全班成績(array 也可以是二維)

```python
import numpy as np
import matplotlib.pyplot as plt
gradesAll = np.array([[85,70,82],[75,98,65],[85,96,96]])
weight = np.array([0.2,0.2,0.6])
np.dot(gradesAll,weight)
```
