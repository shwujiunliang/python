### 線性回歸 scikit-learn


```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
# y = 1.2x +0.8
x = np.linspace(0,5,50)
y = 1.2*x + 0.8
plt.plot(x,y,'r')
# 加入noise randn標準正態分佈
y = 1.2*x + 0.8 + 0.6*np.random.randn(50)
plt.scatter(x,y)
```

```python
x
```

```python
y
```

```python
#調整資料型態 [x,x,x,x,x]=>[[x],[x],[x],[x],[x]]
```

```python
from sklearn.linear_model import LinearRegression
regr = LinearRegression()
X = x.reshape(50,1)
regr.fit(X,y)
Y= regr.predict(X)
```

```python
# 原始的點
plt.scatter(x,y)
#預測的線性回歸
plt.plot(x,Y,'r')
#原始的參數
plt.plot(x,1.2*x+0.8,'g')
```

### 訓練資料和測試資料
#### 取其中的資料作訓練 剩下的作為測試，確認model是否適合

```python
x = np.linspace(0, 5, 100)
y = 1.2*x + 0.8 + 0.5*np.random.randn(100)
plt.scatter(x, y)
```

```python
from sklearn.model_selection import train_test_split
# 把原來的 x, y 中的 80% 給 training data, 20% 給 testing data。
x_train, x_test, y_train, y_test = train_test_split(x, y,
                                                   test_size=0.2,
                                                   random_state=87)
```

### 預測波士頓房價

```python
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
```

```python
#看到數據中所有的 features
boston.feature_names
```

```python
X = boston.data
X
```

```python
Y = boston.target
Y
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y,
                                                   test_size=0.3,
                                                   random_state=87)
regr.fit(x_train, y_train)
y_predict = regr.predict(x_test)
```

```python
plt.scatter(y_test, y_predict)
plt.plot([0,50],[0,50],'r')
plt.xlabel('True Price')
plt.ylabel('Predicted Price')
```

```python
plt.figure(figsize=(8,10))
for i, feature in enumerate(boston.feature_names):
    plt.subplot(5, 3, i+1)
    plt.scatter(X[:,i], Y, s=1)
    plt.ylabel("price")
    plt.xlabel(feature)
    plt.tight_layout()
```
